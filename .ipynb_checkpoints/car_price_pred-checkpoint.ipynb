```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

X=df.drop('Price',axis=1)
y=df['Price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=433)
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
encode_col=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder="passthrough")
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(encode_col,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
ypred=pipe.predict(X_test)
ypred
```

```python
pipe.score(X_test,y_test)
```

```python
score_pipe=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    pipe.fit(X_train,y_train)
    score_pipe.append(pipe.score(X_test,y_test))

max_pipe_sc=max(score_pipe)
print(max_pipe_sc)
print(score_pipe.index(max_pipe_sc))
```
